```python
import pandas as pd
import numpy as np
import seaborn as sns
from pathlib import Path
```

```python
path = Path("../../csv/")
```

```python
titanic = pd.read_csv(path /"titanic_data.csv").dropna()
titanic
```

```python
titanic.head()
```

What is the total number of passengers from each class who survived?

```python
survived_people_from_each_class= titanic.groupby(['class']).agg({
    'survived':'sum'
}).reset_index()
survived_people_from_each_class
```

Question: What was the average fare for passengers from each town?



```python
avg_fare_from_each_town = titanic.groupby(['embark_town']).agg(
    {
        'fare':'mean'
    }
).reset_index()
avg_fare_from_each_town
```

What was the average fare for male vs female passengers from each town?

```python
avg_fare_from_each_town_male_and_female = titanic.groupby(['embark_town','sex']).agg(
    {
        'fare':'mean'
    }
).reset_index()
avg_fare_from_each_town_male_and_female
```

What was the average fare and age for male vs female passengers from each town?

```python
avg_fare_from_each_town_with_age_fare_foreach_sex = titanic.groupby(['embark_town','sex']).agg(
    {
        'fare':'mean',
        'age':'mean'
    }
).reset_index()
avg_fare_from_each_town_with_age_fare_foreach_sex
```

apply set on multiple column

```python
titanic.groupby(['embark_town', 'who'])['age'].apply(set).reset_index()
```

Question: Get the range (min - max) of ages for each age category (who column) from each town.

to apply muliplt agg method on the same column use this -  'age':['min','max']


```python
titanic.groupby(['embark_town','who']).agg(
    min_age = ('age','min'),
    max_age = ('age','max'),
    age_diff = ('age',lambda x:x.max()-x.min())
).reset_index()

```

Question: Get the mean fare by age ratio for each age category (who column) from each town.

```python
titanic.groupby(['embark_town','who']).agg(
    fare_ratio_by_age_ratio = (lambda x: x['fare']/x['age']),
    fare_mean = ('fare','mean')
).reset_index()
```
