```python
import pandas as pd
import numpy as np
```

```python
# Sample log data
logs = pd.DataFrame({
    'timestamp': pd.date_range('2025-01-01', periods=100, freq='H'),
    'user_id': np.random.randint(1, 11, 100),
    'action': np.random.choice(['login', 'view', 'purchase'], 100),
    'revenue': np.random.choice([0, 0, 0, 10, 25, 50], 100)  # Most actions = $0
})
```


* Group by date (extract from timestamp) and action type
* Calculate: count of events, unique users, total revenue
* Add a column for "conversion_rate" (purchases / total events) per day
* Sort by date descending
* Format revenue with 2 decimals

```python
# date
logs['logs_date'] = logs['timestamp'].dt.date
logs.head()
```

```python
logs.groupby(['logs_date','action'])
list(logs)
```

```python
logs['is_purchase'] = (logs['action'] == 'purchase').astype(int)
```

```python
list(logs)
```

Mental model (remember this)

# size = “How many rows?”
# count = “How many non-null values?”

```python
daily_action_metrics = (
    logs.groupby(['logs_date', 'action'])
        .agg(
            event_count=('action', 'size'),
            unique_viewer=('user_id', 'nunique'),
            total_revenue=('revenue', 'sum'),
            purchases=('is_purchase', 'sum')
        )
        .reset_index()
)
```

```python
daily_action_metrics['conversion_rate'] = (
    daily_action_metrics['purchases'] /
    daily_action_metrics['event_count']
)
```
