```python
import pandas as pd
from pathlib import Path
```

```python
path = Path("../../csv/")
```

```python
df = pd.read_csv(path /"data_duplicate.csv")
df
```

```python
#  checking on particular column
print('FULL DATA')
print(df)
print("DUPLCATED DATA")
df[df.duplicated(subset=['order_id'])]
```

using keep 

```python
print("DUPLCATED DATA - first")
# first one is original mark all as duplcaite
df[df.duplicated(subset=['order_id'],keep='first')]
```

```python
print("DUPLCATED DATA - last")
# mark last one is original and  mark all  other as duplcaite
df[df.duplicated(subset=['order_id'],keep='last')]
```

```python
print("DUPLCATED DATA - false")
# mark all as duplcaite
df[df.duplicated(subset=['order_id'],keep=False)]
```
