````python
import pandas as pd
from pathlib import Path
````

````python
path = Path("../../csv/")
````

````python
#  getting data
winter=pd.read_csv(path /"winter.csv",usecols=["Athlete","Medal"])
summer=pd.read_csv(path /"summer.csv",usecols=["Athlete","Medal"])
````

#### practising only merge

````python
print(len(summer),len(winter))
````

````python
winter.merge(summer,how="inner",on="Athlete")
````

````python
#  instead of suffix x and y for column name same on both the table use suffixes to give custom column name it will be appended to the column name
winter.merge(
    summer,
    how="inner",
    on="Athlete",
    suffixes=("_winter", "_summer")
)
````

#### we can also rename the column before merge 

````python
winter_renamed = winter.rename(columns={"Medal": "Winter_Medal"})
summer_renamed = summer.rename(columns={"Medal": "Summer_Medal"})

winter_renamed.merge(
    summer_renamed,
    on="Athlete",
    how="inner"
)
````

````python
winter_renamed.merge(summer_renamed,how="outer",on="Athlete")
````

````python
outer_with_indicator = winter_renamed.merge(summer_renamed,how="outer",on="Athlete",indicator=True)
````

````python
outer_with_indicator.value_counts()
````

We have to remove the both containing values in order to get the outer join excluding intersectionÂ¶

````python
value_both = outer_with_indicator[outer_with_indicator['_merge'] == 'both']
value_both
````

````python
not_value_both = outer_with_indicator[outer_with_indicator['_merge'] != 'both']
not_value_both
````

Using different columns / indexes to join

````python
#  getting data
winter=pd.read_csv(path /"winter.csv",usecols=["Athlete","Medal"])
summer=pd.read_csv(path /"summer.csv",usecols=["Athlete","Medal"])
````

````python
winter_new = winter.rename(columns={"Athlete":"Name"})
````

- left_on -> column from the left table we are joining on 
- right on -> column from the right table we are joining on

````python
final_new = winter_new.merge(summer,how='inner',left_on='Name',right_on='Athlete',indicator=True)
final_new = final_new.drop(columns="Athlete")
final_new
````

concat data

````python
# Scenario: Daily files that need combining
day1 = pd.DataFrame({
    'timestamp': pd.date_range('2025-01-01', periods=3, freq='H'),
    'metric': [100, 105, 110]
})

day2 = pd.DataFrame({
    'timestamp': pd.date_range('2025-01-02', periods=3, freq='H'),
    'metric': [115, 120, 125]
})

day3 = pd.DataFrame({
    'timestamp': pd.date_range('2025-01-03', periods=3, freq='H'),
    'metric': [130, 135, 140]
})

# Stack vertically (append rows)
all_days = pd.concat([day1, day2, day3], ignore_index=True)
print("\nConcatenated daily data:")
print(all_days)

# Add source tracking
day1['source'] = 'day1'
day2['source'] = 'day2'
day3['source'] = 'day3'
tracked = pd.concat([day1, day2, day3], ignore_index=True)
print("\nWith source tracking:")
print(tracked.tail())
````
