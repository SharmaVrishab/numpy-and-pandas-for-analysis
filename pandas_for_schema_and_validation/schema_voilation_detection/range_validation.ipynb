```python
import pandas as pd
```

```python
df = pd.DataFrame({
    "raw_age": ["25", "-5", "abc", None],
    "raw_salary": ["50000", "1000000", "", "xyz"],
    "raw_join_date": ["2024-01-10", "1990-01-01", "2024-02-30", None]
})
```

```python
df = df.convert_dtypes()
df
```

```python
df["age"] = pd.to_numeric(df["raw_age"], errors="coerce")
df["salary"] = pd.to_numeric(df["raw_salary"], errors="coerce")
df["join_date"] = pd.to_datetime(df["raw_join_date"], errors="coerce")
```

```python
df
```

```python
#  missing values from original data
missing_age = df['raw_age'].isna()
missing_salary = df['raw_salary'].isna()
missing_join_date = df['raw_join_date'].isna()
```

```python
# checking for invalid data
invalid_age = df['raw_age'].notna() & df['age'].isna()
invalid_salary = df['raw_salary'].notna() & df['salary'].isna()
invalid_join_date = df['raw_join_date'].notna() & df['join_date'].isna()
```
