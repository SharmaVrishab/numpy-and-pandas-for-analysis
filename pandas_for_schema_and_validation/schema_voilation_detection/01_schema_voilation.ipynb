```python
import pandas as pd
import numpy as np
```

# missing value detection
# isna(), isnull(), notna(), notnull() -> for checking missing value

```python
#  basic type checking

df = pd.DataFrame({
    'age': [25, 30, 'invalid', 35],
    'salary': [50000, 60000, 70000, 80000]
})
print()
# Method 1: Check types match expectations
expected_types = {'age': 'int64', 'salary': 'int64'}
type_violations = df.dtypes != pd.Series(expected_types)
type_violations
```

```python
#  missing value detection
df = pd.DataFrame({
    "age": [25, None, 30, pd.NA],
    "salary": [50000, 60000, None, 70000]
})

df.isna()
```

# ROW AND COLUMN WISE DETECTION

# The axis parameter is what determines the direction:

<!-- 
    axis=0 (default): operates column-wise (down the rows)
    axis=1: operates row-wise (across the columns)
 -->

```python
df = pd.DataFrame({
    'A': [1, 2, np.nan, 4],
    'B': [5, np.nan, np.nan, 8],
    'C': [9, 10, 11, 12]
})

df
```

```python
# Detect missing values (returns boolean DataFrame)
df.isna()  # or df.isnull()
```

# COLUMN WISE DETECTION

```python
# Count missing values per column
# default axis=0
print(df.isna().sum())

# Check if any column has missing values
# per column
print(df.isna().any())

# Check if all values in a column are missing
print(df.isna().all())
```

# ROW WISE DETECTION

```python
df
```

```python
# Count missing values per row
print(df.isna().sum(axis=1))

# Check if any row has missing values
print(df.isna().any(axis=1))

# Check if all values in a row are missing
print(df.isna().all(axis=1))

# Get rows with at least one missing value
print(df[df.isna().any(axis=1)])
```
