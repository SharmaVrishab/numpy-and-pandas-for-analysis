```python
import pandas as pd
```

```python
df = pd.DataFrame({
    'date_str': ['2024-01-01', '2024-13-45', '2024-03-15', 'invalid']
})

# Detect invalid dates
df['date_parsed'] = pd.to_datetime(df['date_str'], errors='coerce')
invalid_dates = df['date_parsed'].isna() & df['date_str'].notna()

if invalid_dates.any():
    print("Invalid dates found:")
    print(df.loc[invalid_dates, 'date_str'])
```
