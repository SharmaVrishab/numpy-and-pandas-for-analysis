```python
import pandas as pd
```

```python
# Single date
pd.to_datetime('2024-12-25')
pd.to_datetime('12/25/2024', format='%m/%d/%Y')
```

```python
#  there is distinction on scaler vs column values
print(type(pd.to_datetime("12/25/2024")))
# pandas.Timestamp

#  here we are using index
print(type(pd.to_datetime(['2024-01-01', '2024-06-15', '2024-12-31'])))
# DatetimeIndex

#  here we are using series or dataframe
s = pd.to_datetime(pd.Series(["2024-12-25", "2024-12-26"]))
print(type(s))
# pandas.DatetimeIndex
print(s.dtype)




```

**timestamp is how dates are stored in pandas and internally they are stored as datetime64ns tha is numpy object not python**
