## practising extracting things from date component

```python
# Demo DataFrame
data = {
    "id": [1, 2, 3, 4, 5],
    "date_str": ["2024-01-15", "2024-02-20", "2024-04-25", "2024-12-31","2024-12-12"],
    "event": ["A", "B", "C", "D", "E"]
}

df = pd.DataFrame(data)
```

```python
#  checking the dtype of dates
df.dtypes
```

```python
df = df.convert_dtypes()
```

```python
df.dtypes
```

```python
df['date_data']= pd.to_datetime(df['date_str'])
df
```

```python
df.dtypes
```

```python
#  creating a large amount of dates

df = pd.DataFrame({
    "date": pd.date_range(start="2024-01-01", periods=30, freq="D"),
    "sales": range(100, 130)
})
df
```

```python
df['date'].dtype
```

```python
df['year'] = df['date'].dt.year
df['month'] = df['date'].dt.month
df['month_name'] = df['date'].dt.month_name()
df['date_extract'] = df['date'].dt.day

df
```

```python

df = pd.DataFrame({
    "raw_date": [
        "2024-01-01",
        "2024-02-30",   # invalid day
        "not_a_date",   # invalid string
        "2024-13-01",   # invalid month
        "",  # empty date
        None            # missing
    ]
})
```

```python
df["parsed_date"] = pd.to_datetime(df["raw_date"], errors="coerce")
df
```

```python
df['parsed_date']
```

```python
df["parsed_date"].isna()
```
