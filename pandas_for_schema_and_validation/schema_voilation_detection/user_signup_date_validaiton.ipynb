```python
import pandas as pd
```

```python
df = pd.DataFrame({
    "user_id": range(1, 11),
    "signup_date_raw": [
        "2024-01-10",   # valid
        "2024/01/15",   # valid but different separator
        "2024-02-30",   # invalid day
        "2024-13-01",   # invalid month
        "",             # empty string
        None,           # missing
        "   ",          # whitespace
        "2023-12-31",   # valid but before allowed year
        "01-02-2024",   # ambiguous format
        "not a date"    # garbage
    ]
})
```

```python
df
```

```python
parsed_dates = pd.to_datetime(df['signup_date_raw'],errors='coerce')
parsed_dates
```

```python
# missing -> none , nan , nat -> these means absent not error
# invalid ->   if a value exists but cannot be a real date.
missing_mask = df['signup_date_raw'].isna()
missing_mask
```

```python
invalid_mask = parsed_dates.isna() & df['signup_date_raw'].notna()
invalid_mask
```

```python
df["date_status"] = "valid"

df.loc[missing_mask, "date_status"] = "missing"
df.loc[invalid_mask, "date_status"] = "invalid"
df
```
