```python
"""LIST **** FLEXIBLE AND MUTABLE"""

my_list = [1,2,3,4,5]
print(my_list)

generated_l = list(range(1,6))
print(generated_l)

# accessing
print(generated_l[0])
print(generated_l[-1])

nested_list = [1,2,3,[4,5,6],[[7,8,9]]]
print(nested_list)
print(nested_list[4][0])

# changing the values
nested_list[4][0] = 'changed'
print(nested_list)


# adding to the list

adding_list = [1,2,3,4,5]
adding_list.append(6)
adding_list.append([7,8,9])
adding_list.extend([10,9,8])


adding_list.insert(6,0)
print(adding_list)
```

```python
# REMOVING
r_list = [1,2,3,4,5]
#  delete at an index
removed = r_list.pop(3)
#  delete by value
r_list.remove(3)
#  this is not list specific but how obj are deleted in python
del r_list[0]
#  clear the full list
r_list.clear()
r_list
```

```python
#mutating list

m_list = list(range(1,10))
print("original list : ",m_list)
m_list.reverse()
print(m_list)
m_list.sort()
print(m_list)
m_list.sort(reverse=True)
print(m_list)
mutated_list = sorted(m_list)
print(mutated_list)

```

```python
# getting info
info_list = [3,4,5,5,5,6,7,7]
# occurance of elements in that list
print(info_list.count(5))
# check for length of the script
print(len(info_list))
#  at index for the first occurance
print(info_list.index(7))


# copying the list
original_l = list(range(5,13))
second = original_l
second.append(12)
print(original_l)

#  we can also use copy which returns a copy of array
#  so original array is not affected
copy_1 = original_l.copy()
copy_1.append([1,2,3])
print(original_l)


#  we can also use slice which returns a copy
#  so original array is not affected
copy_2 = original_l[:]
copy_2.append([1,2,3])
print(original_l)
# we can use pop without an index too which will pop the last index
original_l.pop()

print(original_l)
```

```python
#  above we shallow copy but that will not affect nested list
nested_list = [1,2,3,[4,5,6],[[7,8,9]]]
copy_3 = nested_list[:]
print(copy_3)

#  let try to change the deep nested one
copy_3[4][0][1] = 'changed'
#  so the original list is affected even though we are workig on  slice list which creates a copy but
# it is a shallow copy
print("original list - ",nested_list)
print(copy_3)

```
