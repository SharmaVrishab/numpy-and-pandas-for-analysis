```python
def yes_outer_change():
    outer_var = 'astronut'
    def inner():
        nonlocal outer_var
        outer_var = 'fire truck'
        inner_var = 'spy'
        print(' i wanted to be a ',outer_var)
    inner()
    print("most kid wanted to be a firefighter")
    print("but i wanted to be an" , outer_var)
    try:
        print("but being a ",inner_var,' could be coool')
    except:
        print(
            'too secret for me'
        )
```

```python
yes_outer_change()
```
