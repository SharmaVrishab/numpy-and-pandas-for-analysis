****numpy has structured array which can bascially store multiple dtype

```python
import numpy as np

# Define a structured dtype
dtype = [
    ('name', 'U10'),
    ('age', 'i4'),
    ('weight', 'f4')
    ]

# Create a structured array
s_arr = np.array(
    [
            ('Alice', 25, 55.5),
            ('Bob', 30, 75.0),
            ('Charlie', 22, 68.2)
    ],
    dtype=dtype)

print(s_arr)
```

```python
#  we can convert this strcutured array to dataframe
data = pd.DataFrame(s_arr)
data
```
